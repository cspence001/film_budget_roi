```python
# Dependencies
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Google API Key
from config import api_key
```

```python
#import csv
movies_df = pd.read_csv("Resources/MovieData.csv")
movies_df.head(20)
```

```python
movies_df['Country'] = ""
movies_df['Date Released'] = ""
movies_df['Production Company'] = ""
movies_df['Critic rating'] = ""

movies_df.head()
```

```python
params = {"key": api_key}

for index, row in movies_df.iterrows():
    base_url = "http://www.omdbapi.com/?apikey=" + api_key + "&t=" + 'movie'
    
    movie_name = row['movie_name']
    
    #update movie name key variable
    params['movie'] = movie_name
    
    print(f"Retrieving Results for Index {index}: {movie_name}.")
    
    #make request
    response = requests.get(base_url, params = params).json()
    print(response)
    

```
